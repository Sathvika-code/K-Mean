```python
 from enum import auto
 import numpy as np
 from sklearn.datasets import load_breast_cancer
 from sklearn.decomposition import PCA
 from sklearn.cluster import KMeans
 import matplotlib.pyplot as plt
 from sklearn.cluster import KMeans
 data = load_breast_cancer().data
 pca = PCA(2)
 df = pca.fit_transform(data)
 kmeans = KMeans(n_clusters=10)
 label=kmeans.fit_predict(df)
 cluster_centers = kmeans.cluster_centers_
 plt.figure(figsize=(8, 6))
 for cluster_id in range(10):
    cluster_data = df[label == cluster_id]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster_id+1}')
 plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=100, c='black', label='Cluster Centers')
 plt.title('KMeans Clustering')
 plt.xlabel('Feature 1')
 plt.ylabel('Feature 2')
 plt.legend()
 plt.grid(True)
 plt.show()
```

```python
 from enum import auto
 import numpy as np
 from sklearn.datasets import load_digits
 from sklearn.decomposition import PCA
 import matplotlib.pyplot as plt
 from sklearn.cluster import DBSCAN
 from sklearn.datasets import make_blobs
 data = load_digits().data
 pca = PCA(2)
 df = pca.fit_transform(data)
 dbscan = DBSCAN(eps=1.5, min_samples=10)
 label = dbscan.fit_predict(df)
 colors=['blue','red','green','yellow','pink','orange','black','brown','maroon','gray']
 for i in range(10):
    plt.scatter(df[label == i,0], df[label ==i, 1], label=i, c=colors[i])
 plt.title("DBSCAN Clustering")
 plt.xlabel("Feature 1")
 plt.ylabel("Feature 2")
 plt.legend(loc="upper right")
 plt.show()
```
